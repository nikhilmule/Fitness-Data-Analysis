```python
q5.rename({'SleepDay': 'ActivityDate'}, axis=1, inplace=True)
q4.rename({'ActivityDay': 'ActivityDate'}, axis=1, inplace=True)
q3.rename({'ActivityDay': 'ActivityDate'}, axis=1, inplace=True)
q2.rename({'ActivityDay': 'ActivityDate'}, axis=1, inplace=True)
```

```python
n_d = pd.concat([q1, q2, q3, q4, q5], axis=0, ignore_index=True)
```

```python
n_d
```

```python
n_q = n_d.sort_values(by = "ActivityDate")
```

```python
n_q
```

```python
# here dayfirst is used to align the date column in second position without distubing the format.
# converting object type to datetime type

n_q["Activity_date"] = pd.to_datetime(n_q["ActivityDate"], dayfirst = True)
```

```python
n_q
```

```python
q_d = n_q.sort_values(by = "Activity_date")
```
